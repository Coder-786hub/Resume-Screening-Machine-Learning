Import Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Import clean Dataset

```python
clean_df = pd.read_csv("clean_resume_data.csv")
```

```python
clean_df.head()
```

Exploratory Data Analysis

```python
plt.figure(figsize = (15,5))
sns.countplot(clean_df["Category"])
plt.xticks(rotation = 90)
plt.show()
```

```python
counts = clean_df["Category"].value_counts()
labels = clean_df["Category"].unique()
plt.figure(figsize = (15,10))
plt.pie(counts,labels = labels, autopct = "%1.1f%%", shadow = True, colors = plt.cm.plasma(np.linspace(0,1,3)))
plt.show()
```

Balanced Datasets

```python
from sklearn.utils import resample

# Define the maximum count among all categories
max_count = clean_df["Category"].value_counts().max()

# Resample each category to match the maximum count 
balanced_data = []
for category in clean_df["Category"].unique():
    category_data = clean_df[clean_df["Category"] == category]
    if len(category_data) < max_count:
        #  perform oversampling for categories with fever samples
        balanced_category_data = resample(category_data, replace = True, n_samples = max_count, random_state = 42)
    else:
        # perform undersampling for categories with more samples
        balanced_category_data = resample(category_data, replace = True, n_samples = max_count, random_state = 42)
    balanced_data.append(balanced_category_data)

# Concatenate the balanced data for all categories
balanced_df = pd.concat(balanced_data)
```

```python
balanced_df["Category"].value_counts()
```

Train-Test-Split

```python
balanced_df.isnull().sum()
# balanced_df.dropna(inplace = True)
```

```python
x = balanced_df["Feature"]
y = balanced_df["Category"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

Encoding( TF-IDF)

```python
tfidf_vectorizer = TfidfVectorizer()
x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf = tfidf_vectorizer.transform(x_test)
```

Train Random Forest Classifier

```python

rf_classifier =RandomForestClassifier()
rf_classifier.fit(x_train_tfidf,y_train)

# step 4.: Accuracy Evaluation
y_pred = rf_classifier.predict(x_test_tfidf)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy", accuracy)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Confusion Matrics
conf_metrix = confusion_matrix(y_test,y_pred)
plt.figure(figsize = (10,8))
sns.heatmap(conf_metrix, annot = True, fmt = "d", cmap = "Blues", xticklabels = rf_classifier.classes_, yticklabels = rf_classifier.classes_)
plt.xlabel("Predected")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

Predictive System

```python
import re

def cleanResume(txt):
    cleanText = re.sub(r"http\s+", " ", txt)  # raw string for regex
    cleanText = re.sub(r"RT|cc", " ", cleanText)
    cleanText = re.sub(r"#\s+", " ", cleanText)
    cleanText = re.sub(r"@\s+", " ", cleanText)
    # Use raw string to avoid escape sequence issues
    cleanText = re.sub(r"[%s]" % re.escape(r"""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), " ", cleanText)
    cleanText = re.sub(r"[^\x00-\x7f]", " ", cleanText)
    cleanText = re.sub(r"\s+", " ", cleanText)
    return cleanText


# Predection Function
def predict_category(resume_text):
    resume_text = cleanResume(resume_text)
    resume_tfidf = tfidf_vectorizer.transform([resume_text])
    predicted_category = rf_classifier.predict(resume_tfidf)[0]
    return predicted_category
```
